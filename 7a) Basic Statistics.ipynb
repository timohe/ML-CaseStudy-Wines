Based on the data above there are several features which distinguish redwine from whitewine e.g. residual sugar seems to be significantly different (white wine is indeed sweeter) as well as volatile acidity and sulfur dioxide values. 
I assume that a clustering algorithm is able to pick up those differences, maybe dropping some features would be of help.

```python
redwinestats = redwinedata.describe()
whitewinestats = whitewinedata.describe()
```

### T-Tests
Because it will also be useful for 7.c) we calculate some T-Test. 
The results show that all of the differences in the features are significant.
In order of descending absolute t-values, the significance is:
1. total sulfur dioxide 
2. volatile acidity
3. chlorides
4. sulvates
5. fixed acidity
6. free sulfur dioxide
7. density
8. residual sugar
9. pH
10. citric acid
11. quality
12. alcohol

```python
for featureNr in range(0, redwinestats.shape[1]):
    print(list(redwinestats)[featureNr] + ": ")
    print(
    ttest_ind_from_stats(
        mean1=redwinestats.iloc[1,featureNr], std1=redwinestats.iloc[2,featureNr], nobs1=redwinestats.iloc[0,featureNr], 
        mean2=whitewinestats.iloc[1,featureNr], std2=whitewinestats.iloc[2,featureNr], nobs2=whitewinestats.iloc[0,featureNr]
    )
    )
    print()
```
